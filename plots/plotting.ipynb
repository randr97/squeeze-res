### Comparision of our models with the bottleneck models

```python
'''
The below plot shows how our model with Squeeze and Excitation outperforms the bottleneck blocks.
Also we see that our Mish Activation functions perform better than ReLU activation functions in general.
'''
plt.plot(squeeze_rez_test_loss, label='Resnet With SE')
plt.plot(resnet_bottleneck_mish["testing_loss"], label='Resnet-Mish (Bottleneck) without SE')
plt.plot(resnet_bottleneck_relu["testing_loss"], label='Resnet-RELU (Bottleneck) without SE')
plt.title('Test Loss Comparision')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.savefig('resent__mish__relu__se__loss')
plt.show()
```

```python
plt.plot(squeeze_rez_test_acc, label='Resnet With SE')
plt.plot(resnet_bottleneck_mish["testing_acc"], label='Resnet-Mish (Bottleneck) without SE')
plt.plot(resnet_bottleneck_relu["testing_acc"], label='Resnet-RELU (Bottleneck) without SE')
plt.title('Test Accuracy Comparision')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.savefig('resent__mish__relu__se__acc')
plt.show()
```
